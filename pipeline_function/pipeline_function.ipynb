```python
from transformers import pipeline

# sentiment-analysis is the task
classifier = pipeline("sentiment-analysis")

classifier([
    "I have been waiting for a Hugging Face course my whole life",
    "I hate this too much"
])
```

2 - multiple classification

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification")
classifier(
    "This is the code about the transformer library",
    candidate_labels=["education", "politics", "business"],
)
```

4 - Question Answering

```python
from transformers import pipeline

question_answere = pipeline("question-answering")
question_answere(
    question = 'where do i work?',
    context = 'my name is kavinda and i work in primed pharmacy'
)
```
